```python
# clase con 4 atritubos: id, fecha, region, estado
class Region:
    
    def __init__(self, id, fecha, region, estado):
        self._id = id
        self._fecha = fecha
        self._region = region
        self._estado= estado
            
    @property
    def dni(self):
        return self._fecha
    
    @property
    def nombre(self):
        return self._region
    
    @property
    def nombre(self):
        return self._estado
    
    def __str__(self):
        return f"id: {self.id}, fecha: {self.fecha}, region: {self.region}, estado: {self.estado}"
```

```python
class RegionInexistenteException(Exception):
    pass
```

```python
def buscar_por_region_id(region_id, df):
    res = df[df['RegionID'] == region_id]
    if len(res) == 0: 
        raise RegionInexistenteException('No existe region')
    return res 
```

```python
buscar_por_region_id(13480, zillow_df)
```
